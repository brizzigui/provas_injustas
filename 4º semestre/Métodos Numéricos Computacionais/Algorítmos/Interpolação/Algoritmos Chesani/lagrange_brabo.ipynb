```python
def lagrange(x, y, z):
  m = len(x);

  r = 0;
  for i in range(m):
    c = d = 1;
    for j in range(m):
      if i != j:
        c *= (z-x[j]);
        d *= (x[i] - x[j]);
    r += y[i] * c/d;
  return r;
```

```python
def f(x):
  return 5*x**3 - 2*x**2 - x + 3; #exemplo 3.7 do livro

x = [0.0, 0.2, 0.3, 0.4, 0.7, 0.9]; 
y = [round(f(i), 3) for i in x];
z = 0.6;
```

```python
r = lagrange(x, y, z);

print('valor método:       ', round(r, 6));
print('valor do polinomio: ', f(z));
print('erro =              ', abs(r-f(z)));
```
