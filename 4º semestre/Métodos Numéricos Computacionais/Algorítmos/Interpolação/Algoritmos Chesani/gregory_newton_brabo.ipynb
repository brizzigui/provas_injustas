```python
def gregory_newton(x, y, z): #conjunto de pontos x, conjunto de imagens y, ponto a ser interpolado z
  m = len(x);
  del_y = y.copy();x
  
  for i in range(m):
    for j in range(i+1, m)[::-1]: #[::-1] inverte uma lista, equivalente a range(m-1, i, -1)
      del_y[j] -= del_y[j-1];
      
  u = (z-x[0])/(x[1]-x[0]); #(x-x0)/h
  r = del_y[-1];

  for i in range(m-1)[::-1]:
    r = r * (u-i)/(i+1) + del_y[i];

  return r;
```

```python

def f(x):
  return 4*x**6 + (3/2)*x**4 + 4*x**2 + 17; #expressão qualquer pra testar o erro

## exemplo 1 do livro
#x = [0.1, 0.6];
#y = [1.221, 3.320];
#z = 0.2;

#exemplo 2 do livro
#x = [110, 120, 130];
#y = [2.041, 2.079, 2.114];
#z = 115;


x = [round((i*0.1), 1) for i in range(5)]; #range(n) pontos com (i*gap)
y = [f(i) for i in x]; #conjunto imagem de x
z = 0.5; #ponto a ser interpolado
```

```python
print('valor método       = ', gregory_newton(x,y,z));
print('valor do polinomio = ', f(z));
print('erro               = ', abs(gregory_newton(x,y,z)-f(z)));
```
