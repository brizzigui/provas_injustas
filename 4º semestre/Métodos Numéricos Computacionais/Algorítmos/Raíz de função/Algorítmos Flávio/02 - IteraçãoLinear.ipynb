# Método da Iteração Linear

Dado uma equação f(x) = 0, reescrevemos ela da forma x = g(x), isolando um x de um dos lados.

Note que o ponto em que x = g(x) é a raiz de f(x).

Digamos que $$f(x) = x^2 - x - 2$$

Isolando o x em f(x) = 0, temo que $$ x = \sqrt{x + 2}$$

E, finalmente, analisando os gráficos:

```python
x = symbols('x')

def g(x):
    return sqrt(x + 2)
    
plot(x, g(x), (x, 0, 5))
```

Pela análise gráfica percebemos que a raiz está próxima do valor 2.

Para dar inicio ao método, pegamos um valor inicial para o chute, digamos, 5.

$$g(5) = \sqrt{7} = 2,645$$

Como queremos encontrar o ponto em que g(x) = x, faremos com que o nosso próximo x seja igual a 2,645.

De maneira geral:

$$x_{k+1} = g(x_k)$$

# Ordem de convergência

Se temos que o erro na iteração K é:

$$e_k = | x_k - \overline{x} |$$

Se existe p tal que

$$ \lim_{k\to\infty} \frac {|e_{k+1}|} {|e_k|^p} > 0 $$

p é dito a ordem de convergencia do método

Ou então, para um k suficientemente grande:

$$ p = \frac {\ln{ \frac {e_{k+1}} {e_k}}} {\ln{ \frac{e_k} {e_{k-1}} }} $$

```python
#com controle máximo do número de iterações
def itLinMethod(f, ini, err):
    x = f(ini)
    mx = 100
    
    for _ in range(mx):
        x2 = f(x)
        
        if abs(x2 - x) < err:
            return f(x2)
        
        x = x2
        
    print("Número máximo do iterações, ", mx ,", excedido")
```

```python
from math import *
def func(x):
    return sqrt(x+2)

print(itLinMethod(func, 4, 0.001))
```
