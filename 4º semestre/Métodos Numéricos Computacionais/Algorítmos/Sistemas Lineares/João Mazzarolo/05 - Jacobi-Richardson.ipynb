```python
SL = [[5.0, 2.0, 1.0, 0.0], [3.0, 5.0, 1.0, -7.0], [1.0, 1.0, 3.0, -5.0]]

for i in SL:
    for j in i:
        print("[{0:^5}]".format(j), end="")
    print()
```

```python
n = len(SL)
A = []

for i in range(n):
    linha = []
    for j in range(n + 1):
        linha.append(SL[i][j])
    A.append(linha)

b = []
x = []
v = []

for i in range(n):
    b.append(A[i][n])   
    x.append(0)         # inicializando x com valores de 0 para uso no futuro
    v.append(0)         # inicializando v com valores de 0 para uso no futuro

tol = 0.00001
itermax = 50
CondErro = 0
```

```python
for i in range(n):
    r = 1 / A[i][i]
    for j in range(n):
        if(i != j):
            A[i][j] = A[i][j] * r
    b[i] = b[i] * r
    x[i] = b[i]

Iter = 0
NormRel = 1

while Iter < itermax:
    Iter += 1
    for i in range(n):
        soma = 0
        for j in range(n):
            if i != j:
                soma += A[i][j] * x[j]
        v[i] = b[i] - soma
    NormNum = 0
    NormDen = 0
    for i in range(n):
        t = abs(v[i] - x[i])
        NormNum += t * t
        NormDen += v[i] * v[i]
        x[i] = v[i]
    NormRel = NormNum / NormDen
if NormRel <= tol:
    CondErro = 0
else:
    CondErro = 1
```
