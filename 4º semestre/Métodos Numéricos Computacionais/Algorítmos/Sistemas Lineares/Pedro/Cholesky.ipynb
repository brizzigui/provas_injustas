```python
def Cholesky(A):
    n = len(A);
    erro = 0;
    det = 1;
    for j in range(n):
        soma = 0;
        for k in range(j):
            soma += A[j][k] ** 2;
        t = A[j][j] - soma;
        if t > 0:
            A[j][j] = t**(1/2);
            r = 1/A[j][j];
            det *= t;
        else:
            erro = 1;
        for i in range(j+1, n):
            soma = 0;
            for k in range(j):
                soma += A[i][k] * A[j][k];
            A[i][j] = (A[i][j] - soma) * r;
    return A, det, erro;

        
def SolveTriangSup(LL, b, n):
    y=[]
    for l in range(0,n):
        y.append(0)
        
    y[0]=b[0]/LL[0][0]
    for i in range (1,n):
        soma=0.0
        for j in range(0,i):
            soma=(soma+LL[i][j]*y[j])
        y[i]=(b[i]-soma)/LL[i][i]
    
    x=[]
    for l in range(0,n):
        x.append(0)
        
    x[n-1]=y[n-1]/LL[n-1][n-1]
    for i in range(n-2,-1,-1):
        soma=0.0
        for j in range(i+1,n):
            soma+=[i][j]*x[j]
        x[i]=(y[i]-soma)/Amp[i][i]
    
    return y, x


def EspelhaMatriz(LL, n):
    for i in range(n):
        for j in range(n):
            if (j > i):
                LL[i][j] = LL[j][i]


```

```python
n = 4
m = 5
A = [
        [6 , 2, 6.5, 8, -2.256418], 
        [2, 6.5, 8, 18.125, 3.6804975], 
        [6.5, 8, 18.125, 32, 1.54519225], 
        [8, 18.125, 32, 66.03125, 7.654655625]
      ]

b = [-2.256418, 3.6804975, 1.54519225, 7.654655625]

print('=' * 10)
print(' ENTRADA')
print('=' * 10)
print('A matriz A para a decomposição é:')
for l in range(0, n):
    for c in range (0, m):
        print("[{0:^8}]".format(Amp[l][c]), end='')
    print()
print()

LL, det, erro = Cholesky(A)

EspelhaMatriz(LL, n)

print('=' * 10)
print(' SAÍDA')
print('=' * 10)
print('Matriz Espelhada:')
print()
for l in range(0,n):
    for c in range(0,n):
        print("[{0:^20}]".format(LL[l][c]), end='')
    print()
print()

y, x = SolveTriangSup(LL, b, n)

print('Solução do Sistema Triangular Inferior')
print('- y -')
for l in range(0, n):
    print("[{0:^5}]\n".format(y[l]),end='')
print()
print('Solução do Sistema Triangular Superior')
print('- x -')
for l in range(0, n):
    print("[{0:^5}]\n".format(x[l]),end='')
print()


```
